```python
# Import libraries
import pandas as pd
import os
import shutil
```

```python
meta_data=pd.read_csv('E:\\online_competitions_data\\DL#+Beginner\\DL# Beginner\\meta-data\\meta-data\\train.csv')
```

```python
meta_data.head()
meta_data_train=meta_data[0:10400]
meta_data_test=meta_data[10400:]
```

```python
animal_list=meta_data['Animal'].unique()
```

```python
# Creating seperate folder  
def animal_folder_creation(folder_name, animal_list):
    for animal in animal_list:
        animal_folder_path="E:\\online_competitions_data\\IdentifyAnimal\\dataset\\"+folder_name+"\\"+animal
        if not os.path.exists(animal_folder_path):
            os.makedirs(animal_folder_path, 777)
```

```python
# Mapping images to the correct folder
def mapping_images(meta_data,train_test):
    for index, row in meta_data.iterrows():
        print(row)
        src="E:\\online_competitions_data\\DL#+Beginner\\DL# Beginner\\train\\train\\"+row['Image_id']
        dest="E:\\online_competitions_data\\IdentifyAnimal\\dataset\\"+train_test+"\\"+row["Animal"]
        #print(dest)
        shutil.copy(src,dest)
```

```python
animal_folder_creation('train',animal_list)
animal_folder_creation('test',animal_list)
mapping_images(meta_data_train,'train')
mapping_images(meta_data_test,'test')
```
